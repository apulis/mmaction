```python
import json
import numpy as np
```

```python
with open('..\\..\\activity_net.v1-3.min.json') as f:
  data = json.load(f)

intervals = []

database = data['database']
for key in database:
    anns = database[key]['annotations']
    for ann in anns:
        segments = ann['segment']
        interval = segments[1] - segments[0]
        intervals.append(interval)
        #if interval <= 0.1:
        #    print(segments[0], segments[1])
        
print('segments #:', len(intervals))
```

```python
inp = np.asarray(intervals)
print('segment interval: min={}, max={}'.format(inp.min(), inp.max()))
```

```python
intv_small = inp[np.where( inp < 0.5 )]
print(len(intv_small))
```

```python
%matplotlib notebook

import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

num_bins = 50
n, bins, patches = plt.hist(intv_small, num_bins, facecolor='blue', alpha=0.2)
plt.show()
```
